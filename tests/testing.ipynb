```python
import sqlite3, json
db="../data/recipes.db"
conn=sqlite3.connect(db)
cur=conn.cursor()
print("recipes:", cur.execute("SELECT id,title,source_url,servings,updated_at,version FROM recipes").fetchall())
print("ingredients:", cur.execute("SELECT id,name,default_unit FROM ingredients").fetchall())
print("recipe_ingredients:", cur.execute("SELECT id,recipe_id,ingredient_id,qty_raw,uom,qty_gml FROM recipe_ingredients").fetchall())
print("changes_log (last 10):", cur.execute("SELECT id,entity,entity_id,op,version,updated_at FROM changes_log ORDER BY id DESC LIMIT 10").fetchall())
conn.close()
```
